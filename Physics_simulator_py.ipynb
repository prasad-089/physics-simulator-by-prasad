```python
# ==============================
# PHYSICS SIMULATOR v1.0
# Created by Prasad
# ==============================

import matplotlib.pyplot as plt
import math
import numpy as np

def get_float_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Invalid input! Please enter a number.\n")

def get_positive_input(prompt):
    while True:
        value = get_float_input(prompt)
        if value > 0:
            return value
        else:
            print("Value must be positive. Please try again.\n")
# ------------------------------
# Utility
# ------------------------------
def line():
    print("=" * 40)

# ------------------------------
# Physics Functions
# ------------------------------
def momentum():
    m = get_positive_input("Enter mass (kg): ")
    v = get_float_input("Enter velocity (m/s): ")
    p = m * v

    print(f"\nMomentum = {p:.2f} kg·m/s")


    if p < 10:
        print("Explanation: This is low momentum, so the object is easy to stop.\n")
    elif p < 100:
        print("Explanation: This is moderate momentum. It will require some force to stop.\n")
    else:
        print("Explanation: This is high momentum. The object will be difficult to stop.\n")
def power():
    f = get_float_input("Enter force (N): ")
    v = get_float_input("Enter velocity (m/s): ")
    p = f * v
    print(f"\nPower = {p:.2f} W")
    print("Explanation: Power is the rate of doing work.\n")

def potential_energy():
    m = get_positive_input("Enter mass (kg): ")
    h = get_positive_input("Enter height (m): ")
    g = 9.8
    pe = m * g * h
    print(f"\nPotential Energy = {pe:.2f} J")
    print("Explanation: Stored energy due to height.\n")

def kinetic_energy():
    m = get_positive_input("Enter mass (kg): ")
    v = get_float_input("Enter velocity (m/s): ")
    ke = 0.5 * m * v**2
    print(f"\nKinetic Energy = {ke:.2f} J")
    print("Explanation: Energy due to motion.\n")

def kinematics_solver():
    print("\n--- KINEMATICS SOLVER (SUVAT) ---")

    print("Choose what you want to find:")
    print("1. Final velocity (v)")
    print("2. Initial velocity (u)")
    print("3. Acceleration (a)")
    print("4. Time (t)")
    print("5. Displacement (s)")

    choice = input("Enter choice: ")

    if choice == "1":
        u = get_float_input("Enter initial velocity (u): ")
        a = get_float_input("Enter acceleration (a): ")
        t = get_positive_input("Enter time (t): ")
        v = u + a*t
        print(f"\nFinal velocity v = {v:.2f} m/s")
        print("Used equation: v = u + at\n")

    elif choice == "2":
        v = get_float_input("Enter final velocity (v): ")
        a = get_float_input("Enter acceleration (a): ")
        t = get_positive_input("Enter time (t): ")
        u = v - a*t
        print(f"\nInitial velocity u = {u:.2f} m/s")
        print("Used equation: u = v - at\n")

    elif choice == "3":
        u = get_float_input("Enter initial velocity (u): ")
        v = get_float_input("Enter final velocity (v): ")
        t = get_positive_input("Enter time (t): ")
        a = (v - u) / t
        print(f"\nAcceleration a = {a:.2f} m/s²")
        print("Used equation: a = (v - u) / t\n")

    elif choice == "4":
        u = get_float_input("Enter initial velocity (u): ")
        v = get_float_input("Enter final velocity (v): ")
        a = get_float_input("Enter acceleration (a): ")
        t = (v - u) / a
        print(f"\nTime t = {t:.2f} s")
        print("Used equation: t = (v - u) / a\n")

    elif choice == "5":
        u = get_float_input("Enter initial velocity (u): ")
        a = get_float_input("Enter acceleration (a): ")
        t = get_positive_input("Enter time (t): ")
        s = u*t + 0.5*a*t**2
        print(f"\nDisplacement s = {s:.2f} m")
        print("Used equation: s = ut + ½at²\n")

    else:
        print("Invalid choice.\n")
# ------------------------------
# Graph Generator (mobile-friendly)
# ------------------------------

def graph_motion():
    print("\n--- Motion Graph Generator ---")

    u = get_float_input("Enter initial velocity u (m/s): ")
    a = get_float_input("Enter acceleration a (m/s²): ")
    t_max = int(get_positive_input("Enter total time (max 300): "))

    if t_max > 300:
        print("Too large, using 300 instead.")
        t_max = 300

    time = list(range(t_max + 1))

    velocity = [u + a * t for t in time]
    displacement = [u * t + 0.5 * a * t**2 for t in time]

    print("\nChoose graph:")
    print("1. Velocity vs Time")
    print("2. Displacement vs Time")
    print("3. Both graphs")

    choice = input("Enter choice: ")

    plt.close('all')


    if choice == "1":
        plt.figure()
        plt.plot(time, velocity)
        plt.xlabel("Time (s)")
        plt.ylabel("Velocity (m/s)")
        plt.title("Velocity vs Time")
        plt.grid(True)
        plt.savefig("velocity_graph.png")
        plt.show()
        plt.close()

    elif choice == "2":
        plt.figure()
        plt.plot(time, displacement)
        plt.xlabel("Time (s)")
        plt.ylabel("Displacement (m)")
        plt.title("Displacement vs Time")
        plt.grid(True)
        plt.savefig("displacement_graph.png")
        plt.show()
        plt.close()

    elif choice == "3":
        plt.figure()

        plt.subplot(2, 1, 1)
        plt.plot(time, velocity)
        plt.title("Velocity vs Time")
        plt.xlabel("Time (s)")
        plt.ylabel("Velocity (m/s)")
        plt.grid(True)

        plt.subplot(2, 1, 2)
        plt.plot(time, displacement)
        plt.title("Displacement vs Time")
        plt.xlabel("Time (s)")
        plt.ylabel("Displacement (m)")
        plt.grid(True)

        plt.tight_layout()
        plt.savefig("both_graphs.png")
        plt.show()
        plt.close()
    else:
        print("Invalid choice.\n")

def plot_trajectory(u, theta, g):
    t_flight = (2 * u * math.sin(theta)) / g

    t_values = [i * t_flight / 100 for i in range(101)]

    x = [u * math.cos(theta) * t for t in t_values]
    y = [u * math.sin(theta) * t - 0.5 * g * t**2 for t in t_values]

    plt.close('all')
    plt.figure()
    plt.plot(x, y)
    plt.xlabel("Horizontal Distance (m)")
    plt.ylabel("Vertical Height (m)")
    plt.title("Projectile Trajectory")
    plt.grid(True)

    plt.savefig("projectile.png")
    plt.close()

    print("\nTrajectory saved as projectile.png")
    print("Open left panel → Files → projectile.png\n")

def projectile_motion():
    print("\n--- PROJECTILE MOTION ---")

    u = get_float_input("Enter initial velocity u (m/s): ")
    angle = get_float_input("Enter angle of projection θ (degrees): ")

    g = 9.8

    # Convert angle to radians
    theta = math.radians(angle)

    # Formulas
    time_of_flight = (2 * u * math.sin(theta)) / g
    max_height = (u**2 * (math.sin(theta))**2) / (2 * g)
    range_proj = (u**2 * math.sin(2 * theta)) / g

    print(f"\nTime of Flight = {time_of_flight:.2f} s")
    print(f"Maximum Height = {max_height:.2f} m")
    print(f"Range = {range_proj:.2f} m")

    print("\nExplanation:")
    print("- Time of flight depends on vertical component of velocity.")
    print("- Maximum height occurs when vertical velocity becomes zero.")
    print("- Range depends on sin(2θ).\n")

    # Ask if user wants trajectory graph
    choice = input("Do you want trajectory graph? (y/n): ")

    if choice.lower() == "y":
        plot_trajectory(u, theta, g)

def work_energy_theorem():
    print("\n--- WORK–ENERGY THEOREM ---")

    m = get_positive_input("Enter mass (kg): ")
    u = get_float_input("Enter initial velocity u (m/s): ")
    v = get_float_input("Enter final velocity v (m/s): ")

    ke_initial = 0.5 * m * u**2
    ke_final = 0.5 * m * v**2

    work_done = ke_final - ke_initial

    print(f"\nInitial Kinetic Energy = {ke_initial:.2f} J")
    print(f"Final Kinetic Energy = {ke_final:.2f} J")
    print(f"Work Done (W = ΔK) = {work_done:.2f} J")


    if work_done > 0:
        print("Explanation: Positive work done → object gains kinetic energy (speeds up).\n")
    elif work_done < 0:
        print("Explanation: Negative work done → object loses kinetic energy (slows down).\n")
    else:
        print("Explanation: Zero work done → kinetic energy remains constant.\n")

# Universal Gravitational Constant
G = 6.67e-11

# Planetary Presets (All major bodies)
planets = {
    "earth": {"M": 5.972e24, "R": 6.371e6},
    "moon": {"M": 7.347e22, "R": 1.737e6},
    "mars": {"M": 6.417e23, "R": 3.389e6},
    "jupiter": {"M": 1.898e27, "R": 6.991e7},
    "saturn": {"M": 5.683e26, "R": 5.823e7},
    "uranus": {"M": 8.681e25, "R": 2.536e7},
    "neptune": {"M": 1.024e26, "R": 2.462e7},
    "mercury": {"M": 3.285e23, "R": 2.440e6},
    "venus": {"M": 4.867e24, "R": 6.052e6},
    "sun": {"M": 1.989e30, "R": 6.963e8}
}

def gravitation_and_satellites():
    while True:
        print("\n--- GRAVITATION & SATELLITES TOOL ---")
        print("1. Gravitational Force")
        print("2. Gravitational Field")
        print("3. Escape Velocity")
        print("4. Orbital Velocity")
        print("5. Time Period of Satellite")
        print("6. Earth Preset (g, escape velocity)")
        print("7. Geostationary Satellite (Earth)")
        print("8. Variation of g with height (Earth)")
        print("9. Variation of g with depth (Earth)")
        print("10. Gravitational Potential Energy")
        print("11. All Planet Presets")
        print("12. Plot graph of g vs height (Earth)")
        print("13. Exit")

        choice = input("Choose option: ")

        if choice == "1":
            m1 = get_positive_input("Mass m1 (kg): ")
            m2 = get_positive_input("Mass m2 (kg): ")
            r = get_positive_input("Distance (m): ")
            F = G * m1 * m2 / r**2
            print(f"Force = {F:.3e} N")

        elif choice == "2":
            M = get_positive_input("Mass of planet (kg): ")
            r = get_positive_input("Distance (m): ")
            g = G * M / r**2
            print(f"Field strength = {g:.3e} N/kg")

        elif choice == "3":
            M = get_positive_input("Mass (kg): ")
            R = get_positive_input("Radius (m): ")
            ve = math.sqrt(2 * G * M / R)
            print(f"Escape Velocity = {ve:.2f} m/s")

        elif choice == "4":
            M = get_positive_input("Mass (kg): ")
            r = get_positive_input("Orbital radius (m): ")
            v = math.sqrt(G * M / r) # Corrected: used math.sqrt and removed get_float_input around calculation
            print(f"Orbital Velocity = {v:.2f} m/s")

        elif choice == "5":
            M = get_positive_input("Mass (kg): ")
            r = get_positive_input("Orbital radius (m): ")
            T = 2 * math.pi * math.sqrt(r**3 / (G * M))
            print(f"Time Period = {T:.2f} s")

        elif choice == "6":
            M = planets["earth"]["M"]
            R = planets["earth"]["R"]
            g = G * M / R**2
            ve = math.sqrt(2 * G * M / R)

            print("\n--- Earth Preset ---")
            print(f"Mass = {M:.3e} kg")
            print(f"Radius = {R:.3e} m")
            print(f"Surface gravity = {g:.2f} m/s²")
            print(f"Escape velocity = {ve:.2f} m/s")

        elif choice == "7":
            M = planets["earth"]["M"]
            R = planets["earth"]["R"]
            T = 24 * 3600

            r = (G * M * T**2 / (4 * math.pi**2)) ** (1/3)
            h = r - R

            print(f"Orbital radius = {r:.2f} m")
            print(f"Height above Earth = {h/1000:.2f} km")

        elif choice == "8":
            h = get_positive_input("Height above Earth (m): ")
            M = planets["earth"]["M"]
            R = planets["earth"]["R"]
            g = G * M / (R + h)**2
            print(f"g at height = {g:.4f} m/s²")

        elif choice == "9":
            d = get_positive_input("Depth below Earth (m): ")
            M = planets["earth"]["M"]
            R = planets["earth"]["R"]
            g = G * M * (R - d) / R**3
            print(f"g at depth = {g:.4f} m/s²")

        elif choice == "10":
            m = get_positive_input("Object mass (kg): ")
            M = get_positive_input("Planet mass (kg): ")
            r = get_positive_input("Distance from center (m): ")
            U = -G * M * m / r
            print(f"Gravitational Potential Energy = {U:.3e} J")

        elif choice == "11":
            print("\nAvailable celestial bodies:")
            for name in planets:
                print("-", name.capitalize())

            body = input("Choose one: ").lower()

            if body in planets:
                M = planets[body]["M"]
                R = planets[body]["R"]
                g = G * M / R**2
                ve = math.sqrt(2 * G * M / R)

                print(f"\n--- {body.capitalize()} ---")
                print(f"Mass = {M:.3e} kg")
                print(f"Radius = {R:.3e} m")
                print(f"Surface gravity = {g:.2f} m/s²")
                print(f"Escape velocity = {ve:.2f} m/s")
            else:
                print("Invalid name.")

        elif choice == "12":
            M = planets["earth"]["M"]
            R = planets["earth"]["R"]

            h = np.linspace(0, 5e6, 200)   # heights from 0 to 5000 km
            g = G * M / (R + h)**2

            plt.figure()
            plt.plot(h/1000, g)
            plt.xlabel("Height above Earth (km)")
            plt.ylabel("g (m/s²)")
            plt.title("Variation of g with Height")
            plt.grid(True)
            plt.savefig("g_vs_height.png")
            plt.show()
            plt.close()

        elif choice == "13":
            print("Exiting Gravitation Tool.")
            break

        else:
            print("Invalid choice.")


# ------------------------------
# Main Menu
# ------------------------------
def main():
    while True:
        line()
        print("      PHYSICS SIMULATOR v1.0")
        print("          Created by Prasad")
        line()

        print("1. Momentum")
        print("2. Power")
        print("3. Potential Energy")
        print("4. Kinetic Energy")
        print("5. Motion Graph Generator")
        print("6. Kinematics Solver")
        print("7. Projectile Motion")
        print("8. Work–Energy Theorem")
        print("9. Gravitation & Satellites")
        print("10. Exit")
        choice = input("\nChoose an option: ")

        line()

        if choice == "1":
            momentum()
        elif choice == "2":
            power()
        elif choice == "3":
            potential_energy()
        elif choice == "4":
            kinetic_energy()
        elif choice == "5":
            graph_motion()
        elif choice == "6":
            kinematics_solver()
        elif choice == "7":
            projectile_motion()
        elif choice == "8":
            work_energy_theorem()
        elif choice == "9":
            gravitation_and_satellites()
        elif choice == "10":
            print("Thank you for using Physics Simulator!")
            break
        else:
            print("Invalid choice. Try again.\n")

# Run program
main()
```
